```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("../data/raw/sales_data.csv")

# Display Dataset
print("========== FIRST 5 RECORDS ==========")
print(df.head())

# Dataset Information
print("\n========== DATASET INFO ==========")
print(df.info())

# Missing Values
print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

# Remove Duplicates
df.drop_duplicates(inplace=True)

# Convert Date
df["Date"] = pd.to_datetime(df["Date"])

# Calculate Final Amount
df["Final_Amount"] = (
    df["Quantity"] *
    df["Price"] *
    (1 - df["Discount"] / 100)
)

# Total Revenue
print("\n========== TOTAL REVENUE ==========")
print(f"₹{df['Final_Amount'].sum():,.2f}")

# Average Order Value
print("\n========== AVERAGE ORDER VALUE ==========")
print(f"₹{df['Final_Amount'].mean():,.2f}")

# Monthly Sales
df["Month"] = df["Date"].dt.strftime("%b")

monthly_sales = df.groupby("Month")["Final_Amount"].sum()

print("\n========== MONTHLY SALES ==========")
print(monthly_sales)

# Top Products
top_products = (
    df.groupby("Product")["Final_Amount"]
      .sum()
      .sort_values(ascending=False)
)

print("\n========== TOP PRODUCTS ==========")
print(top_products)

# Category Sales
category_sales = (
    df.groupby("Category")["Final_Amount"]
      .sum()
      .sort_values(ascending=False)
)

print("\n========== CATEGORY SALES ==========")
print(category_sales)

# City Sales
city_sales = (
    df.groupby("City")["Final_Amount"]
      .sum()
      .sort_values(ascending=False)
)

print("\n========== CITY SALES ==========")
print(city_sales)

# Payment Analysis
payment_mode = df["Payment_Mode"].value_counts()

print("\n========== PAYMENT MODE ==========")
print(payment_mode)

# ---------------- Graph 1 ----------------

plt.figure(figsize=(8,5))
monthly_sales.plot(kind="bar")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()

# ---------------- Graph 2 ----------------

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")
plt.title("Category Sales")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()

# ---------------- Graph 3 ----------------

plt.figure(figsize=(7,7))
payment_mode.plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.title("Payment Mode Distribution")
plt.tight_layout()
plt.show()

print("\nProject Completed Successfully!")
```
