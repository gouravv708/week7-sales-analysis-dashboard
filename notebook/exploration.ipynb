```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("../data/raw/sales_data.csv")

print("=" * 50)
print("SALES DATASET EXPLORATION")
print("=" * 50)

# First 5 Records
print("\nFirst 5 Records")
print(df.head())

# Last 5 Records
print("\nLast 5 Records")
print(df.tail())

# Dataset Shape
print("\nDataset Shape")
print(df.shape)

# Column Names
print("\nColumns")
print(df.columns)

# Dataset Information
print("\nDataset Information")
df.info()

# Summary Statistics
print("\nSummary Statistics")
print(df.describe())

# Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Duplicate Records
print("\nDuplicate Records")
print(df.duplicated().sum())

# Convert Date Column
df["Date"] = pd.to_datetime(df["Date"])

# Revenue Calculation
df["Final_Amount"] = (
    df["Quantity"] *
    df["Price"] *
    (1 - df["Discount"] / 100)
)

# Revenue Preview
print("\nRevenue Preview")
print(df[["Product", "Quantity", "Price", "Discount", "Final_Amount"]].head())

# Product Count
print("\nProducts")
print(df["Product"].value_counts())

# Category Count
print("\nCategories")
print(df["Category"].value_counts())

# City Count
print("\nCities")
print(df["City"].value_counts())

# Payment Modes
print("\nPayment Modes")
print(df["Payment_Mode"].value_counts())

# Histogram
plt.figure(figsize=(8,5))
plt.hist(df["Final_Amount"], bins=10)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

print("\nExploration Completed Successfully!")
```
